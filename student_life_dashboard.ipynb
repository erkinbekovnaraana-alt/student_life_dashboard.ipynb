```python
import pandas as pd
import plotly.express as px

# создаем данные
data = {
    'study_hours': [1, 2, 3, 4, 5, 6, 7, 8],
    'phone_hours': [8, 7, 6, 5, 4, 3, 2, 1],
    'sleep_hours': [5, 6, 6, 7, 7, 8, 8, 9],
    'exam_score': [50, 55, 60, 65, 70, 75, 85, 90]
}

df = pd.DataFrame(data)

# график 1: учеба vs оценки
fig1 = px.scatter(df, x='study_hours', y='exam_score',
                  title='Study Hours vs Exam Score')
fig1.show()

# график 2: сон vs оценки
fig2 = px.scatter(df, x='sleep_hours', y='exam_score',
                  title='Sleep vs Exam Score')
fig2.show()

# график 3: телефон vs оценки
fig3 = px.scatter(df, x='phone_hours', y='exam_score',
                  title='Phone Usage vs Exam Score')
fig3.show()
```

## Conclusion

The analysis shows that study time and sleep have a positive impact on exam performance,
while excessive phone usage negatively affects student success.
